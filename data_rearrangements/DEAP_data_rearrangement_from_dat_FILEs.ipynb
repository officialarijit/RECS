```python
import pandas as pd
import numpy as np
import glob
import pickle
import scipy.io
import os
```

```python
#=================================================
#Folder creation for rearranged DEAP dataset
#=================================================

datapath = '/DEAP Data/data_preprocessed_python/'

deapEEG = 'DEAP_eeg/' #Folder name
deapEDA = 'DEAP_eda/' #Folder name
deapRB = 'DEAP_respbelt/' #Folder name

pathEEG = os.path.join(datapath,deapEEG)
pathEDA = os.path.join(datapath,deapEDA)
pathRB = os.path.join(datapath,deapRB)
os.mkdir(pathEEG)
os.mkdir(pathEDA)
os.mkdir(pathRB)
```

```python
#========================#
# EEG data rearrangement #
#========================#

eeg_channels = 32 #from 1 to 32
videos = 40 #number of videos 
participants = 32 #number of participants
print('==================================================================')
print('Re-arranged EEG data processing')
print('------------------------------------------------------------------')
print('[Persion No, Video No., Channel No., EEG signal, Valence, Arousal]')
print('==================================================================')


for p in range(0,participants): #particioants loop 
    print('-----------------------------------')
    print('Participant:',p+1)
    print('-----------------------------------')
    final_data = []
    if p <10:
        flpath = datapath+'s'+str(0)+str(p+1)+'.dat'
    else:
        flpath = datapath+'s'+str(p+1)+'.dat'
    df = pickle.load(open(flpath,'rb'), encoding='latin1')
    signal = df['data']
    emotion_labels = df['labels']
    for v in range(0,videos): #video loop
        print('------------------')
        print('Video Number:',v+1)
        print('------------------')
        signal_based_video = signal[v,:]
        for ch in range(0,eeg_channels): #eeg channels loop 
            final_data.append(np.concatenate(([p+1],[v+1],[ch+1],signal_based_video[ch,:],emotion_labels[v,[0,1]])))
    
    fdata= pd.DataFrame(final_data)
    fileName = pathEEG+str(p+1)+'_EEG.csv'
    fdata.to_csv(fileName, header=None, index=None)        
```
